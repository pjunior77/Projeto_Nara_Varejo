```python
#Importo as bibliotecas
import pandas as pd
```

```python
#Carrega as bases extras de vendas
dfVendas1 = pd.read_csv("venda 1.csv", delimiter=",")
dfVendas2 = pd.read_csv("venda 2.csv", delimiter=",")
dfVendas3 = pd.read_csv("venda 3.csv", delimiter=",")
```

```python
#Concateno as bases em uma só
dfVendasExtras = pd.concat([dfVendas1, dfVendas2, dfVendas3], ignore_index=True)
dfVendasExtras.head(2)
```

```python
#Transformo a coluna Data para o tipo de dados Data
dfVendasExtras["Data"] = pd.to_datetime(dfVendasExtras["Data"].str.replace(" 00:00:00", "", regex=False))
dfVendasExtras.info()
```

```python
#Transformo a coluna Valor_Total para float
dfVendasExtras["Valor_Total"] = dfVendasExtras["Valor_Total"].astype(float)/100
dfVendasExtras.info()
```

```python
#Adiciono a coluna "ID_Campanha" ao dataFrame
dfVendasExtras["ID_Campanha"] = -1
dfVendasExtras.head(2)
```

```python
#Formato a coluna Valor_Total com duas casas decimais
dfVendasExtras["Valor_Total"] = dfVendasExtras["Valor_Total"].map('{:.2f}'.format)
dfVendasExtras.head(2)
```

```python
#Acerto o valor da coluna ID_Venda, pois o arquivo "vendas 1.csv" tinha Vendas com o mesmo ID
dfVendasExtras["ID_Venda"] = dfVendasExtras["ID_Venda"] + 801
dfVendasExtras.head(2)
```
